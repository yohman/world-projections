```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('worldpop.geojson')
```

```python
# global projection list
projections = [('WGS84','epsg:4326'),
               ('Mercator','epsg:3857'),
               ('Molleweide','esri:54009'),
               ('World Eckert IV','esri:54012'),
               ('World Robinson','esri:54030'),
               ('World Winkel','esri:54018')
              ]
```

```python
fig,ax = plt.subplots(len(projections),1,figsize=(18,8*len(projections)))

for i in range(len(projections)):
    gdf=gdf.to_crs(projections[i][1])
    gdf.plot(ax=ax[i])            
    ax[i].set_title(projections[i][0]+' ('+projections[i][1]+')')
    ax[i].axis('off')
```
